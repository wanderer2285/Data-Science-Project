```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
train_data=pd.read_csv('train.csv')
```

```python
df = train_data
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8, 5))
sb.histplot(df['Age'].dropna(), kde=True, bins=30, color='skyblue')
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Barplot: Gender vs. Survival
plt.figure(figsize=(8, 5))
sb.countplot(x='Sex', hue='Survived', data=df, palette='Set2')
plt.title('Gender vs. Survival')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
# Plot Heatmap for Correlation
plt.figure(figsize=(10, 6))
correlation_matrix = df.corr(numeric_only=True)
sb.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()
```
